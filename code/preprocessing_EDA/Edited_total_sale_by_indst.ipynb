```python
sidos = data["CARD_SIDO_NM"].unique()
stds = data["STD_CLSS_NM"].unique()
```

```python
temp = []

for sido in sidos:
    for std in stds:
        temp.append([sido,std])
```

```python
print(temp)
```

```python
indst = pd.DataFrame(temp)
```

```python
indst.rename(columns = {0 : "CARD_SIDO_NM", 1:"STD_CLSS_NM"}, inplace = True)
```
