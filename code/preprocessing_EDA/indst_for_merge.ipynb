```python
data = pd.read_csv("april_add_merge_with_homeproportion.csv", encoding = "euc-kr")
```

```python
data.drop(["Unnamed: 0", "CSTMR_CNT", "CNT", "AMT"], axis= 1, inplace =True)
```

```python
indst = data[["CARD_SIDO_NM", "STD_CLSS_NM", "total_sale_by_indst"]]
```

```python
grouped = indst.groupby(["CARD_SIDO_NM", "STD_CLSS_NM"]).mean()
```

```python
grouped = grouped.reset_index()
```
